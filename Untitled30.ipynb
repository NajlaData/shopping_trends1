```python
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from IPython.display import Image
import warnings

df = pd.read_csv(r"C:\Users\Najla\Downloads\archive (3)\shopping_trends.csv")
df.head()
```

```python
# تحليل الفئات والتصنيفات
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Category')
plt.title('Distribution of Purchased Items by Category')
plt.show()

# تحليل الأعمار
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', bins=20)
plt.title('Age Distribution of Customers')
plt.show()

# تحليل الجنس
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Gender')
plt.title('Distribution of Customers by Gender')
plt.show()

# تحليل المبالغ المشتراة
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Purchase Amount (USD)', bins=20)
plt.title('Distribution of Purchase Amounts')
plt.show()

# تحليل طرق الدفع
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Payment Method')
plt.title('Distribution of Payment Methods')
plt.xticks(rotation=45)
plt.show()

# تحليل المراجعات والتقييمات
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Review Rating', bins=5)
plt.title('Distribution of Review Ratings')
plt.show()
```
